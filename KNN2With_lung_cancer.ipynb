```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Datasets/lung_cancer_examples.csv")
df
```

```python
#Check if is there any missing value
df.isna().sum()
```

```python
# Check if all the values are in numeric
df.dtypes
```

```python
# In the above we have non-numeric values as object, in the column Name and Surname
# Those values(Name,Surname) not dependent with cancer prediction we drop those Columns
df.drop(["Name","Surname"],axis = 1,inplace = True)
print(df)
```

```python
# we need to take Features seperately as "X"
X = df.iloc[:,:4]   # OR df.iloc[:,:-1]
print(X)
```

```python
# we need to take Target seperately as "y"
y = df.iloc[:,-1]
y
```

```python
# Converting all the features-values to Normalised/Standardised value
# By using Scaling
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
Scaler.fit(X)
X_Scaled = Scaler.transform(X)
X_Scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_Scaled,y,test_size = 0.3,random_state = 11)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
# Model Building
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
# Prediction based on X_test
y_pred = knn.predict(X_test)
y_pred
```

```python
# y_test contains the actual values
y_test.values
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
y_new = knn.predict(Scaler.transform([[78,6,7,6]]))
if y_new.item()==1:
  print("Person will suffer lung cancer")
else:
  print("Person will not suffer lung cancer")
# here the out put will be with an error because the features is in the form dataframes to convert it into tthe form of array the error wil be solved for that when we add features to X use .values with it
```
