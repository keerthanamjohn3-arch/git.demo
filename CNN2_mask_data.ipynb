```python
from skimage.io import imread
from skimage.transform import resize
import os
from keras.models import Sequential
from keras.layers import Conv2D,MaxPool2D,Dense,Flatten
```

```python
train_path = "/content/drive/MyDrive/Maskdata"


X_train = []
y_train = []

categories = os.listdir(train_path)
# print(categories)

for category in categories:
  cat_path = os.path.join(train_path,category)
  print(cat_path)

  images = os.listdir(cat_path)
  # print(images)

  for img in images:
    img_path = os.path.join(cat_path,img)
    image = imread(img_path)

    img_resize = resize(image,(150,150,3))

    X_train.append(img_resize)
    y_train.append(categories.index(category))

  print(category,"Loaded.........")

```

```python
test_path = "/content/drive/MyDrive/Maskdatatest"

X_test = []
y_test = []

for category in categories:
  cat_path = os.path.join(test_path,category)
  images = os.listdir(cat_path)

  for img in images:
    img_path = os.path.join(cat_path,img)
    image = imread(img_path)

    img_resize = resize(image,(150,150,3))
    X_test.append(img_resize)
    y_test.append(categories.index(category))
  print(category,"Loaded.........")
```

```python
len(X_train),len(y_train)
```

```python
len(X_test),len(y_test)
```

```python
import numpy as np
X_train = np.array(X_train)
X_test  = np.array(X_test)
y_test = np.array(y_train)
y_train = np.array(y_test)
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Conv2D(32,(3,3),input_shape =(150,150,3),activation = 'relu'))
model.add(MaxPool2D())
model.add(Flatten())
model.add(Dense(100,activation = 'relu'))
model.add(Dense(2,activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam',loss  = 'sparse_categorical_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(X_train,y_train,epochs = 50,batch_size =12)
```

```python
model.evaluate(X_test,y_test)
```

```python
model.save('mask_data.keras')
```
